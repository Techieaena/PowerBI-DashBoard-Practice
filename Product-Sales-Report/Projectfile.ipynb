```python
import pandas as pd

df = pd.read_excel("Product-Sales-Region.xlsx")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
df['promotion'].fillna('No Promotion', inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['delivery_days'] = (df['deliverydate'] - df['orderdate']).dt.days
```

```python
df.head()
```

```python
df['year'] = df['orderdate'].dt.year
df['month'] = df['orderdate'].dt.month
```

```python
df.head()
```

```python
region_sales = df.groupby('region')['totalprice'].sum().reset_index()
region_sales
```

```python
avg_delivery = df.groupby('region')['delivery_days'].mean().reset_index()
avg_delivery
```

```python
df.to_excel("cleaned_sales_data.xlsx", index=False)
```
